## Ensure annotation example

```python
def test_prod(a: int, b: int) -> int:
    return a * b
```

```python
print(test_prod(2, 3))
print(test_prod(2, "3"))
```

```python
pip install ensure
```

```python
from ensure import ensure_annotations

@ensure_annotations
def test_prod(a: int, b: int) -> int:
    return a * b
```

```python
print(test_prod(2, 3))
print(test_prod(2, "3"))
```
